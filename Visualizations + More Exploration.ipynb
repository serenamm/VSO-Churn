```python
import pandas as pd
import numpy as np

basepath = "E:/vsodataforSerena/"
data = pd.read_csv(basepath + "final_data.csv")
```

```python
data.set_index("Unnamed: 0", inplace = True)
```

```python
data.head()
```

Time to do some data viz! Let's look at each variable one by one.

```python
import matplotlib.pyplot as plt
import matplotlib.style
% matplotlib inline
```

```python
unique_orders = data["Unique_Orders"].value_counts()
unique_orders[unique_orders > 5].plot.barh()
plt.show()
```

```python
data_ = data["Churn?"].value_counts()
data_.plot.barh()
plt.show()
```

```python
moves = data["Num_Moves"].value_counts()
moves.plot.barh()
plt.show()
```

```python
inact = data["inactive_reason"].value_counts()
inact.plot.barh()
plt.show()
```

```python
pc = data["Priority_Code"].value_counts()
pc.plot.barh()
plt.show()
```

```python
rs = data["Recent_Solicitor"].value_counts()
rs_ = rs[rs > 150]
rs_.plot.barh()
plt.show()
# Majority of people either had tickets previously, or bought their tickets online, or from "shawn!"
```

```python
# Who last spent the most?
max_purchase = max(data["Recent_Purchase"])
max_cust = (data[data["Recent_Purchase"] == max_purchase])
max_cust
# Someone in Salt Lake City Utah really likes buying tickets- this was their only purchase though, and they returned half
# Strange!
```

```python
# Who last returned the most?
max_return = max(data["Recent_Return"])
max_return_cust = (data[data["Recent_Return"] == max_return])
max_return_cust
# It was also this Salt Lake City person!
```

```python
# Who has purchased the most?
tot_purch = max(data["Total_Purchased"].dropna())
tot_purch_cust = (data[data["Total_Purchased"] == tot_purch])
tot_purch_cust
# Someone right downtown in Vancouver
```

```python
# Who has returned the most?
tot_ret = max(data["Total_Returned"].dropna())
tot_ret_cust = (data[data["Total_Returned"] == tot_ret])
tot_ret_cust
# Someone right downtown in Vancouver
# Same person in DT Vancouver?
```

```python
# Who has paid the most?
tot_pay = max(data["Total_Paid"].dropna())
tot_pay_cust = (data[data["Total_Paid"] == tot_pay])
tot_pay_cust
# Someone in West Van
```
